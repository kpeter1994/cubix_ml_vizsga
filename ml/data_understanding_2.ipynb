```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('articles.csv')
```

```python
df.dropna()
```

```python
df.drop([
    "id",
    "published",
    "image"
], axis=1, inplace=True)
```

```python
df
```

## Kategoriák összevonása


```python
df['category'] = df['category'].replace({
    'global': 'kulfold',
    'hirtvkulfold': 'kulfold',
    'nagyvilag': 'kulfold',
    'vilag': 'kulfold',
    'celeb': 'bulvar',
    'sztarok': 'bulvar',
    'sztarvilag': 'bulvar',
    'kultur': 'kultura',
    'teve': 'szorakozas',
    'techtud': 'tudomany',
    'itthon': 'belfold',
    'nemzetkozi-gazdasag': 'gazdasag',
    'vilaggazdasag-magyar-gazdasag': 'gazdasag',
})
```

## Kategoriák eltávolítása
Értelmetlen valós információt nem tartalmazó kategoriák letávolítása

```python
df = df[~df['category'].isin(['2023', 'hirek', 'aktualis', 'kozelet', 'ahirtvhirei'])]
```

## Kategoriák szűrése
Csak azok a kategoriák kerüljenek be, amik elérnek egy minimum számosságot

```python
category_counts = df['category'].value_counts()

valid_categories = category_counts[category_counts >= 50].index

df = df[df['category'].isin(valid_categories)]
```

```python
import matplotlib.pyplot as plt

top_categories = df['category'].value_counts().nlargest(20)

plt.figure(figsize=(10, 6))
top_categories.plot(kind='bar')
plt.title('Leggyakoribb kategoriák')
plt.xlabel('Kategória')
plt.ylabel('Darabszám')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
from bs4 import BeautifulSoup

df['summary'] = df['summary'].apply(lambda x: BeautifulSoup(str(x), 'html.parser').get_text())
```

```python
df['text'] = df['title'] + ' ' + df['summary']
df.drop(['summary', 'title'], axis=1, inplace=True)
```

```python
import huspacy
import spacy

huspacy.download()
nlp = spacy.load("hu_core_news_lg")

```

```python
def lemmatize(text):
    doc = nlp(text.lower())
    return ' '.join(
        token.lemma_ for token in doc
        if not token.is_stop and not token.is_punct and token.is_alpha
    )

df['lemmatized_text'] = df['text'].apply(lemmatize)
```

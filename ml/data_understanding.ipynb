```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('articles.csv')
```

```python
df
```

```python
df.drop(["image", "published", "id"], axis=1, inplace=True)
```

```python
df
```

## Kategoriák összevonása

```python
df['category'] = df['category'].replace({
    'global': 'kulfold',
    'hirtvkulfold': 'kulfold',
    'vilag': 'kulfold',
    'celeb': 'bulvar',
    'sztarok': 'bulvar',
    'sztarvilag': 'bulvar',
    'kultur': 'kultura',
    'techtud': 'tudomany',
    'itthon': 'belfold',
    'nemzetkozi-gazdasag': 'gazdasag',
    'vilaggazdasag-magyar-gazdasag': 'gazdasag',
})
```

## Kategoriák eltávolítása

```python
df = df[~df['category'].isin(['test-es-lelek', 'hirek', 'ahirtvhirei'])]
```

```python
# Megszámoljuk a kategóriák előfordulását
category_counts = df['category'].value_counts()

# Kiválasztjuk azokat, amik legalább 30-szor szerepelnek
valid_categories = category_counts[category_counts >= 50].index

# Szűrjük az eredeti DataFrame-et
df = df[df['category'].isin(valid_categories)]
```

```python
import matplotlib.pyplot as plt

top_categories = df['category'].value_counts().nlargest(30)

plt.figure(figsize=(10, 6))
top_categories.plot(kind='bar')
plt.title('Top 10 kategória gyakoriság')
plt.xlabel('Kategória')
plt.ylabel('Darabszám')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
